# loading te data


```python
import pandas as pd
import os
import numpy as np
from pathlib import Path
import re
from bs4 import BeautifulSoup
```

```python
files=['/content/easy_ham/easy_ham','/content/hard_ham/hard_ham','/content/spam_2/spam']
#dont forget to change spam_2/spam_2 to spam_2/spam
temp_data = []

for doc_path in files:

  file_names = os.listdir(doc_path)

  for doc in file_names:

    file_path=os.path.join(doc_path,doc)
    if os.path.isfile(file_path):
      content = open(file_path, encoding='ISO-8859-1')
      temp_data.append({'message':content.read(), 'type':doc_path[-4:]})
      content.close()

data=pd.DataFrame(temp_data)

data.head()
```

```python
print(data.iloc[5].message)
```

```python
from email import message_from_string
def clean_text(email):
  try:
    msg = message_from_string(email)

    body = msg.get_payload(decode=True).decode('utf-8')
    if body is not None:

      patterns = [

          r'[^A-Za-z0-9$ ]',
          r'http\S+',
          r"ALB>",

                ]
      combined_pattern = '|'.join(patterns)
      clean_text = re.sub(combined_pattern,'',body).lower()
      return clean_text
    else:
      return ''
  except Exception as e:
    print(f'the email {e}  are not valid ')
    return ''
```

```python
print( clean_text(data.iloc[18].message))
```

```python
print( data.iloc[18].message)
```

```python
data['message']=data['message'].apply( clean_text )
```

# data encoding


```python
def encoding (data):
  data=data[data['message']!='']
  data.reset_index(drop=True,inplace=True)
  data['type']=data['type'].map({'_ham':1,'spam':2})
  data.sample(frac=1,random_state=30)
  return data
```

```python
encoded_data=encoding(data)
encoded_data
```

```python
from sklearn.model_selection import  train_test_split
from sklearn.feature_extraction.text import  CountVectorizer
X=encoded_data['message']
Y=encoded_data['type']

x_train, X_test , y_train, Y_test=train_test_split(X,Y, test_size=0.2,random_state=30)

vectorizer=CountVectorizer(stop_words='english', lowercase=True)

x_train_v=vectorizer.fit_transform(x_train)
X_test_v=vectorizer.transform(X_test)
```

```python
print(x_train_v.shape)
print(X_test_v.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

RF_model = RandomForestClassifier()

RF_model.fit(x_train_v,y_train)

y_pred = RF_model.predict(X_test_v)
acc = accuracy_score(y_pred, Y_test)
acc
```
